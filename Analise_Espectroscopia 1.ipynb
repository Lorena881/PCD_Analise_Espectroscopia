```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df = pd.read_csv("A650_dados.csv", sep = ";")
df
```

```python
plt.figure()
plt.plot(df['concentracao'],df['A650'], label = "A650", marker  ='o', color = 'b')

plt.title("Espectroscopia do azul de metileno")
plt.xlabel("concentração")
plt.ylabel("Absorbancia")
plt.grid(True, alpha = 0.75)
plt.legend()
plt.show()
```
